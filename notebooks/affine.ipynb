```python
import numpy as np
import nibabel as nib
from pathlib import Path
import scipy.spatial

from cv2 import estimateAffine3D
```

```python
def load_patient_info(patient_id: int):
    path = Path('..') / 'data' / 'segthor_train' / 'train' / ('Patient_%02d/GT.nii.gz' % (patient_id,))
    img = nib.load(path)
    array = np.array(img.dataobj)
    print(array.shape)
    return array

def load_patient_info_2(patient_id: int):
    path = Path('..') / 'data' / 'segthor_train' / 'train' / ('Patient_%02d/GT2.nii.gz' % (patient_id,))
    img = nib.load(path)
    array = np.array(img.dataobj)
    print(array.shape)
    return array

for i in range(1, 41):
    load_patient_info(i)
```

```python
points_messed_up = np.argwhere(load_patient_info(27) == 2)
points_correct = np.argwhere(load_patient_info_2(27) == 2)
```

```python
print(points_messed_up.shape, points_correct.shape)
# a = estimateAffine3D(points_messed_up[:1000], points_correct[:1000])
```

```python
def e2h(vec):
    return np.hstack((vec, np.ones((vec.shape[0], 1))))

def h2e(vec):
    return (vec[:,:-1].T / vec[:,-1].T).T

def transform(vec, matrix):
    return h2e(e2h(vec) @ matrix)



temp_transform = np.identity(4)

for iter in range(10):
    # sample points
    n_samples = 10000
    random_messed_up_idx = np.random.choice(np.arange(points_messed_up.shape[0]), size=n_samples)
    random_correct_idx = np.random.choice(np.arange(points_correct.shape[0]), size=n_samples)
    random_messed_up = transform(points_messed_up[random_messed_up_idx], temp_transform)
    random_correct = points_correct[random_correct_idx]

    # move centroids
    t = np.mean(random_messed_up, axis=0) - np.mean(random_correct, axis=0)
    random_messed_up = (random_messed_up - t)
    translation_transform = np.array([[1, 0, 0, t[0]],
                                              [0, 1, 0, t[1]],
                                              [0, 0, 1, t[2]],
                                              [0, 0, 0, 1]])

    # data association
    tree = scipy.spatial.KDTree(random_correct)
    assoc_m = []
    assoc_c = []
    used_idx = set()
    for idx_m, point in enumerate(random_messed_up):
        for idx_c in tree.query(point,k=n_samples)[1]:
            if idx_c not in used_idx:
                used_idx.add(idx_c)
                assoc_m.append(idx_m)
                assoc_c.append(idx_c)
                break

    # estimate transformation
    est = estimateAffine3D(random_messed_up[np.array(assoc_m)], random_correct[np.array(assoc_c)])[1]
    if est is None:
        continue
    est = np.vstack((est, np.array([[0, 0, 0, 1]])))
    print("est=", est)
#     print(est)
    
    # apply transformation i guess?
#     print(temp_transform.shape, translation_transform.shape, est.shape)
    temp_transform = temp_transform @ translation_transform @ np.linalg.inv(est)
    print(temp_transform)
```
